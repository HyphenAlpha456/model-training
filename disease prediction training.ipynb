```python
import joblib as jb
```

```python
max=0;
for i in range(1,1000):
    accuracy=accuracy_score(Y_pred,Y_test)
    if(accuracy>=max):
        max=accuracy
if(max>=0.90):
    print("The accuracy of the model is ",max)
    jb.dump(model,'diseasepred.pkl')
    print("Model saved succesfully")
else:
    print("The accuracy of the model is ",max)
    print("Model not saved")
```
